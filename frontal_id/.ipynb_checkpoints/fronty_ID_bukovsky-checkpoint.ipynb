### import 850 hPa theta_e over the CONUS

```python
import os.path

thetae_850 = []

for file in sorted(glob.glob(os.path.join('/glade/work/doughert/asp/flood_storm_types/era5_*/era5_thetae_*.nc'))):
    print(file)
    file_match = xr.open_dataset(file)
    file_match.close()
    thetae_850.append(file_match)
```

```python
thetae_850_allyears = xr.merge(thetae_850)
```

### determine where front is if thetae gradient > 4 K/ 100 km (Sprenger et al. 2017)
#### or 1.24 K/31 km (ERA5 res)

```python
%%time
import metpy.calc as mpcalc
from metpy.units import units

theta_e_grad = []
fronty = []
frontx = []

for i in range(len(thetae_850_allyears['thetae'])):
    #print(i)
    grad = mpcalc.gradient(thetae_850_allyears['thetae'][i], coordinates=(thetae_850_allyears.latitude, 
                                                                          thetae_850_allyears.longitude))

    # get gradient in x and y directions
    grady = np.abs(grad[0].magnitude)
    gradx = np.abs(grad[1].magnitude)
    
    # nan out values that aren't fronts 
    grady[grady<3.6]= np.nan
    gradx[gradx<3.6]= np.nan
    
    theta_e_grad.append(grad)
    fronty.append(grady)
    frontx.append(gradx)
```

```python
print(len(fronty))
print(np.shape(fronty[0]))
print(np.shape(frontx[0]))
```

#### turn into 0s if nan and 1 if not nan (i.e., a front)

```python
%%time
objecty_id_01s = [fronty_xr[i].notnull().astype(int) for i in range(len(fronty_xr))]
objectx_id_01s = [frontx_xr[i].notnull().astype(int) for i in range(len(frontx_xr))]
```

```python
%%time
timesy = [objecty_id_01s[i].time.values for i in range(len(objecty_id_01s))]
timesx = [objecty_id_01s[i].time.values for i in range(len(objectx_id_01s))]
```

```python
print(len(timesy))
```

### turn object array features into polygons

```python
%%time
import rasterio.features

objecty_id_01s_int = [objecty_id_01s[i].values.astype(rasterio.int32) for i in range(len(objecty_id_01s))]
objectx_id_01s_int = [objectx_id_01s[i].values.astype(rasterio.int32) for i in range(len(objecty_id_01s))]
```

### y front polygons

```python
import rasterio
import shapely

shapes_y = [rasterio.features.shapes(objecty_id_01s_int[i]) for i in range(len(objecty_id_01s))]
```

```python
polygons_y = []

for i in range(len(objecty_id_01s)):
    poly = [shapely.geometry.Polygon(shape[0]["coordinates"][0]) for shape in shapes_y[i] if shape[1] == 1]
    polygons_y.append(poly)
```

```python
print(len(polygons_y))
```

### free up some memory

```python
del thetae_850
del thetae_850_allyears
del theta_e_grad
del fronty_xr
del frontx_xr
del fronty
del frontx
```

#### get polygon coordindates (y front)

```python
%%time
x_coordsy = []
y_coordsy = []
timesy_all = []

for i in range(0,33604):
    for p in range(len(polygons_y[i])):
        x,y = polygons_y[i][p].exterior.coords.xy
        x_coordsy.append(np.array(x).astype(int))
        y_coordsy.append(np.array(y).astype(int))
        timesy_all.append(timesy[i])                                                  
```

```python
print(len(x_coordsy))
print(len(timesy_all))
```

#### get lon,lat corresponding to polygon y coords

```python
%%time
poly_lony = []

for c in range(len(x_coordsy)):
    lon_idx = x_coordsy[c]
    lon_idx[np.where(lon_idx==241)]=240
    lon_vals = objecty_id_01s[0].longitude.values[lon_idx]-360
    poly_lony.append(lon_vals)
```

```python
%%time
poly_laty = []

for c in range(len(y_coordsy)):
    lat_idx = y_coordsy[c]
    lat_vals = objecty_id_01s[0].latitude.values[lat_idx]
    poly_laty.append(lat_vals)
```

```python
%%time
from shapely.geometry import Polygon
polygon_geomy = [Polygon(zip(poly_lony[c], poly_laty[c])) for c in range(len(y_coordsy))]
```

```python
print(len(polygon_geomy))
```

### turn into new polygon with lat/lon coords

```python
geoy_idx = np.arange(0, len(timesy_all), 1)
```

```python
from geopandas import GeoSeries, GeoDataFrame

crs = {'init': 'epsg:4326'}
fronty_polygon = gp.GeoDataFrame(index=geoy_idx, crs=crs, geometry=GeoSeries(polygon_geomy))    
fronty_polygon['time'] = timesy_all
```

### designate region as having frontal influence if frontal object intersects region

```python
%%time
fronty_region = []
fronty_date = []
front_count = []

for f in range(len(conus_reg)):
    for t in range(len(fronty_polygon)):
        if (conus_reg['geometry'].iloc[f].intersects(fronty_polygon['geometry'].iloc[t])):
            fronty_region.append(conus_reg['region_list'].iloc[f])
            fronty_date.append(fronty_polygon['time'].iloc[t])
            front_count.append(1)
            
```

```python
print(len(fronty_region))
```

### make a df of fronts

```python
fronty_df = pd.DataFrame({'region':fronty_region, 'date':fronty_date, 'occurrence':front_count})
```
